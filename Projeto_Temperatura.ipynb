```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
numeros = np.empty(7)

for i in range(0,7):
    numeros[i] = np.random.randint(20,36)

dias = np.array(["Domingo", "Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado"])

```

```python
df = {"Dias": dias, "Temperatura": numeros}

media_temperatura = df['Temperatura'].mean()
maxima = df['Temperatura'].max()
minima = df['Temperatura'].min()

print(f"A temperatura média foi de {media_temperatura:.2f}°")
print(f"A máxima foi de {maxima}° e a mínima foi de {minima}°")
```

```python
x = dias
y = numeros

plt.bar(x,y)

plt.title("Temperatura da Semana")

plt.xlabel("Dias")
plt.ylabel("Temperatura")

plt.show()
```

```python
plt.plot(x,y)

plt.title("Temperatura da Semana")

plt.xlabel("Dias")
plt.ylabel("Temperatura")

plt.show()
```

```python
plt.scatter(x,y)

plt.title("Temperatura da Semana")

plt.xlabel("Dias")
plt.ylabel("Temperatura")

plt.show()
```
