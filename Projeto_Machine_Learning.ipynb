Tratando os dados

```python
df['genre'].unique()
```

```python
mapa = {
 
    'Action': 'Acao',
    'Shooter': 'Acao',
    'Action-Adventure': 'Acao',
    'Fighting': 'Acao',

  
    'Sports': 'Esporte',
    'Racing': 'Esporte',


    'Role-Playing': 'RPG',
    'MMO': 'RPG',


    'Simulation': 'Simulacao',
    'Strategy': 'Simulacao',


    'Misc': 'Casual',
    'Party': 'Casual',
    'Music': 'Casual',


    'Platform': 'Plataforma',
    'Adventure': 'Plataforma',

 
    'Puzzle': 'Puzzle',

 
    'Sandbox': 'Outros',
    'Education': 'Outros',
    'Board Game': 'Outros',
    'Visual Novel': 'Outros'
}
```

```python
df['genre_reduzido'] = df['genre'].map(mapa)

df['genre_reduzido'].value_counts()
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
df['critic_score'] = df['critic_score'].fillna(df['critic_score'].mean())
df['total_sales'] = df['total_sales'].fillna(df['total_sales'].mean())
df['na_sales'] = df['na_sales'].fillna(df['na_sales'].mean())
df['jp_sales'] = df['jp_sales'].fillna(df['jp_sales'].mean())
df['pal_sales'] = df['pal_sales'].fillna(df['pal_sales'].mean())
df['other_sales'] = df['other_sales'].fillna(df['other_sales'].mean())
```

```python
df.sample(5)
```

PREPARANDO PARA O MACHINE LEARNING

```python
import sklearn 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
X = df[['critic_score','total_sales', 'na_sales', 'jp_sales', 'pal_sales', 'other_sales','console']]
y = df['genre_reduzido']

X = pd.get_dummies(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size= 0.2, random_state= 42)
```

```python
#model = LogisticRegression(max_iter = 1000000, class_weight = "balanced")
#model = LinearRegression()
model = RandomForestClassifier(class_weight = "balanced")
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy: ", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
